```python
from skimage.io import imread, imshow
from skimage.util import pad
from imgaug import augmenters as iaa
%matplotlib inline
```

```python
img = imread("../data/images/train/0_0.jpg")
img.shape
```

```python
imshow(img)
```

```python
augm = pad(pad_width=((38, 38), (38, 38), (0, 0)), array=img, mode='maximum')
print("shape = ", augm.shape)
imshow(augm)
```

```python
aug = iaa.Sequential([
    iaa.Fliplr(0.2),
    iaa.Affine(rotate=(-45, 45), mode="edge")
])
```

```python
imshow(aug.augment_image(augm))
```
