```python
import bson
import io
from skimage.io import imread, imshow
import pandas as pd
import seaborn as sns
from tqdm import tqdm
%matplotlib inline
```

```python
data = bson.decode_file_iter(open('../data/train_example.bson', 'rb'))
```

```python
datas = [d for d in data]
```

```python
all_images = []
for item in datas:
    for img in item['imgs']:
        all_images.append(imread(io.BytesIO(img['picture'])))
```

```python
imshow(all_images[7])
```
