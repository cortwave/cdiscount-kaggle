```python
import matplotlib.pyplot as plt
%matplotlib inline
import json
```

```python
steps = []
acc = []
loss = []
with open("../results/resnet50/train_0.log", "r") as f:
    for line in f.readlines():
        if "acc" in line:
            e = json.loads(line)
            steps.append(e['step'])
            acc.append(e['valid_acc'] / 100)
            loss.append(e['valid_loss'])
```

```python
plt.plot(steps, acc)
```

```python
plt.plot(steps, loss)
```
